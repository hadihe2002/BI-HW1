```python
import pandas as pd
import numpy as np
```

```python
def calculate_distance(point1: list, point2: list):
  result = 0
  result += abs(point1[0] - point2[0]) + abs(point1[1] - point2[1])
  return result
```

```python
def create_distance_matrix(points: dict):
  distance_matrix = []
  points_count = len(points)
  for i in range(points_count):
    distance_matrix.append([])
    for j in range(points_count):
      distance_matrix[i].append(calculate_distance(points[i+1], points[j+1]))
  return distance_matrix
```

```python
def beautify_distance_matrix(distance_matrix):
  columns = np.arange(1, len(distance_matrix)+1)
  index = columns
  distance_matrix_np = np.array(distance_matrix)
  distance_df = pd.DataFrame(distance_matrix_np, index=index, columns=columns)
  return distance_df
```

```python
distance_matrix_sorted3 = []
for distance_list in distance_matrix3:
    distance_array_sorted = sorted(distance_list)
    # distance_array_sorted[1:] to eliminate the point itself from the distance list
    distance_matrix_sorted3.append(distance_array_sorted[1:])
np.array(distance_matrix_sorted3)
```

<div style="font-family: Vazirmatn; font-size:24px; direction: rtl; display: flex; align-items:center; justify-content:center; align-text:center; background-color: #4c6ef5; padding: 10px 20px;border: 1px solid black; border-radius: 5px">
محاسبه‌ی میانگین فاصله‌ی K=4 نزدیکترین همسایه
</div>

```python
average_distance_of_nearest_k_neighbors = []
for distance_list in distance_matrix_sorted3:
    average_distance_of_nearest_k_neighbors.append(sum(distance_list[:K3]) / K3) 
np.array(average_distance_of_nearest_k_neighbors)
```

حذف n=5 نقطه‌ی اول
</div>

```python
average_distance_of_nearest_k_neighbors_clone = average_distance_of_nearest_k_neighbors.copy() 
for i in range(n3):
    current_max = max(average_distance_of_nearest_k_neighbors_clone)
    point_number = average_distance_of_nearest_k_neighbors_clone.index(current_max) + 1
    print("Point", point_number, "is a NOISE!")
    average_distance_of_nearest_k_neighbors_clone[point_number-1] = 0
```
